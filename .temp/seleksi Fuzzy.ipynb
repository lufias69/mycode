```python
fuzz_komentar = list()
fuzz_label = list()
ix  = 0
for k,l in zip(komentar_list, label_list):
    print(".", end="")
    if ix%100==0 and ix !=0:
        print(len(fuzz_label))
    input_ = True
    if ix == 0:
        fuzz_komentar.append(k)
        fuzz_label.append(l)
        input_ = False
    else:
        for nn, fk in enumerate(komentar_asli):
            nilai = fuzz.token_sort_ratio(k, fk)
            if nilai > 97:
#                 print(nilai)
                fuzz_komentar.insert(0,k)
                fuzz_label.insert(0,l)
                print("|", end="")
                break
#             if (len(fuzz_komentar)>300 and nn >300):
#                 break
#         if input_ == True:
            
#             print("", end="")
    ix +=1
```
