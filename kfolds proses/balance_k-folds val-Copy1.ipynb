```python
def validasi(X, y, motode, alpha= None):
    if alpha !=None:
        motode.set_params(alpha=alpha)
#     print(motode)
    return motode.score(X,y)

def batches(my_list, n):
    final = [my_list[i * n:(i + 1) * n] for i in range((len(my_list) + n - 1) // n )]
    return final
```
