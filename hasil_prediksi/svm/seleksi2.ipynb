```python
from os import listdir
from os.path import isfile, join
import pandas as pd
import re
```

```python
def dfile (path):
    mypath = path
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    return onlyfiles

def reduksi_huruf(kata):
    nkata  = list()
    for i,k in enumerate(kata):
        if i>2:
            if kata[i]== kata[i-1] and kata[i] == kata[i-2]:
                continue
            else:
                nkata.append(k)
        else:
            nkata.append(k)
    return "".join(nkata)

def hapus_kata(kata, list_kata):
    katas = kata.split()
    for ix, kt in enumerate(katas):
        if kt in list_kata:
            katas[ix]=""
    ret = re.sub(' +', ' '," ".join(katas))
    # save_gdiganti()
    return ret.strip()
```

```python
# nama_file = dfile ('D:\_onedrive\OneDrive\Tesis\implementasi\seleksi_data_latih\data')
alamatt = r"D:\_OneDrive\OneDrive - UGM 365\Penelitian\Koding\hasil_prediksi\svm\new1"
nama_file = dfile (alamatt)
print(nama_file)
apalah = ["ayu","ting","at","rafathar","a", "aa", 'gigi', "raditya", "dika","award","rafatar","raffi", "nagita", "raffinagita1717", "kakak", "kak","baim", "paula","acara","via", "viavallen"]
```

```python
# datataa= pd.read_excel("D:\_onedrive\OneDrive\Tesis\implementasi\seleksi_data_latih\data/hasill_prediksi_remove_kata_.xlsx")
# # k_ = d['komentar'].tolist()
# # l_= d['label'].tolist()
# datataa
```

```python
komentar_list = list()
label_list = list()
ck_list = list()
komentar_kotor = list()
for i in nama_file:
    print(i)
    d= pd.read_excel(alamatt+"/"+i)
    k_ = d['komentar'].tolist()
    k_k = d['komentar_kotor'].tolist()
    l_= d['prediksi'].tolist()
    n_k = list()
    n_l = list()
    tp_l = list()
    n_kk= list()
    
    for k, l, kk in zip (k_, l_, k_k):
        #print(".",end="")
        try:
            k = reduksi_huruf(k)
            k = hapus_kata(k,apalah)
        except:
            continue
        try:
            k_s = k.split()
            if len(k_s) >= 3 and l == "non spam":
                if k not in n_k:
                    n_k.append(k)
                    n_l.append(l)
                    tp_l.append(0) 
                    n_kk.append(kk)
            elif len(k_s) > 5 and l == "spam":
                if k not in n_k:
                    n_k.append(k)
                    n_l.append(l)
                    n_kk.append(kk)
                    for i in apalah:
                        cc = True
                        if i in k_s:
                            tp_l.append(1)
                            cc = False
                            break
                    if cc == True:
                        tp_l.append(0)
        except:
            continue
    #print("")
    komentar_list+=n_k
    label_list +=n_l
    ck_list += tp_l
    komentar_kotor += n_kk
dictt = {
    "label":label_list,
    "komentar":komentar_list,
    "komentar_kotor":komentar_kotor,
    "kode salah": ck_list
}
df = pd.DataFrame.from_dict(dictt)
df.to_excel("hasil_seleksi_New.xlsx")
df
```
