```python
def dfile (path):
    mypath = path
    onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
    return onlyfiles

def reduksi_huruf(kata):
    nkata  = list()
    for i,k in enumerate(kata):
        if i>2:
            if kata[i]== kata[i-1] and kata[i] == kata[i-2]:
                continue
            else:
                nkata.append(k)
        else:
            nkata.append(k)
    return "".join(nkata)

def hapus_kata(kata, list_kata):
    katas = kata.split()
    for ix, kt in enumerate(katas):
        if kt in list_kata:
            katas[ix]=""
    ret = re.sub(' +', ' '," ".join(katas))
    # save_gdiganti()
    return ret.strip()
```

```python
apalah = ["ayu","ting","at","rafathar","a", "aa", 'gigi',
          "raditya", "dika","award","rafatar","raffi", "nagita", 
          "raffinagita1717", "kakak", "kak","baim", "paula","acara",
          "radit",
          "via", "viavallen"]
reduksi_huruf("sayaaaaaa lapaaaaaar beda ituuu")
hapus_kata("gigi raditya dika apalah",apalah)
```

```python
path = r'D:\_OneDrive\OneDrive - UGM 365\Penelitian\Tesis\Koding\spam\hasil_prediksi\uji\latih'
nama_file = dfile (path)
nama_file
```

```python
# import sys
#sys.path.append('D:\_onedrive\OneDrive\Tesis\implementasi\hasil_prediksi')


komentar_list = list()
label_list = list()
ck_list = list()
ixx = 0
for i in nama_file:
    d= pd.read_excel(path+"/"+i)
    k_ = d['komentar'].tolist()
    l_= d['prediksi'].tolist()
    n_k = list()
    n_l = list()
    tp_l = list()
    
    for k, l in zip (k_, l_):
        #print(".",end="")
        if ixx %10 ==0 and ixx !=0 and ixx %100 !=0:
            print(".",end="")
            ixx +=1
        elif ixx %100 ==0:
            print("."+str(ixx))
            ixx +=1
        try:
            k = reduksi_huruf(k)
            k = hapus_kata(k,apalah)
        except:
            continue
        try:
            k_s = k.split()
            if len(k_s) >= 2 and l == "non spam":
                if k not in n_k:
                    n_k.append(k)
                    n_l.append(l)
                    tp_l.append(0)   
            elif len(k_s) >= 3 and l == "spam":
                if k not in n_k:
                    n_k.append(k)
                    n_l.append(l)
                    
                    for i in apalah:
                        cc = True
                        if i in k_s:
                            tp_l.append(1)
                            cc = False
                            break
                    if cc == True:
                        tp_l.append(0)
        except:
            continue
    #print("")
    komentar_list+=n_k
    label_list +=n_l
    ck_list += tp_l
dictt = {
    "label":label_list,
    "komentar":komentar_list,
    "kode salah": ck_list
}
df = pd.DataFrame.from_dict(dictt)
df.to_excel("seleksi/seleksi_all_2.xlsx")
df
```
