```python
import pandas as pd
import numpy as np
```

```python
from Naive_Bayes import naive_bayes as nb
model = nb.NaiveBayesClassifier()
```

```python
data_ = pd.read_excel("data_latih/dataset_bersih_2.xlsx")
latih_cm = np.array(data_['komentar'].tolist())
label_cm = np.array(data_['label'].tolist())
data_
```

```python
model.train(latih_cm, label_cm)
```

```python

model.alpha = 1
pred = model.predict(latih_cm)
# for ix, i in enumerate(latih_cm):
#     pred.append(model.predict(i))
#     if (ix+1)%100==0:
#         print(ix+1)
#     else:
#         print(".",end="")
        
from sklearn.metrics import classification_report
y_true = label_cm
y_pred = pred
target_names = ['non spam', 'spam']
print(classification_report(y_true, y_pred, target_names=target_names))
```
